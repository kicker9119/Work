```python
df= df.drop_duplicates()
df
```

```python
print(df.nunique())
```

```python
unique_JRA = df['JobRoleArea'].unique()
print("Unique values using unique():", unique_JRA)
```

```python
df['JobRoleArea'] = df['JobRoleArea'].replace('InformationTechnology', 'Information_Technology')
df['JobRoleArea'] = df['JobRoleArea'].replace('Information Technology', 'Information_Technology')
df['JobRoleArea'] = df['JobRoleArea'].replace('HumanResources', 'Human_Resources')
df['JobRoleArea'] = df['JobRoleArea'].replace('Human Resources', 'Human_Resources')
```

```python
unique_JRA = df['JobRoleArea'].unique()
print("Unique values using unique():", unique_JRA)
```

```python
unique_Gen = df['Gender'].unique()
print("Unique values using unique():", unique_Gen)
```

```python
unique_Mar = df['MaritalStatus'].unique()
print("Unique values using unique():", unique_Mar)
```

```python
unique_Pay = df['PaycheckMethod'].unique()
print("Unique values using unique():", unique_Pay)
```

```python
df['PaycheckMethod'] = df['PaycheckMethod'].replace('Mail Check', 'Mail_Check')
df['PaycheckMethod'] = df['PaycheckMethod'].replace('Mailed Check', 'Mail_Check')
df['PaycheckMethod'] = df['PaycheckMethod'].replace('MailedCheck', 'Mail_Check')
df['PaycheckMethod'] = df['PaycheckMethod'].replace('DirectDeposit', 'Direct_Deposit')
df['PaycheckMethod'] = df['PaycheckMethod'].replace('Direct Deposit', 'Direct_Deposit')
```

```python
unique_Pay = df['PaycheckMethod'].unique()
print("Unique values using unique():", unique_Pay)
```

```python
unique_NCPW = df['NumCompaniesPreviouslyWorked'].unique()
print("Unique values using unique():", unique_NCPW)
```

```python
df['NumCompaniesPreviouslyWorked'] = df['NumCompaniesPreviouslyWorked'].fillna(0)
```

```python
unique_NCPW = df['NumCompaniesPreviouslyWorked'].unique()
print("Unique values using unique():", unique_NCPW)
```

```python
df["AnnualSalary"]= df["AnnualSalary"].astype("str")
df["AnnualSalary"]= df["AnnualSalary"].str.strip("-")
df
```

```python
df["AnnualSalary"]= df["AnnualSalary"].astype("float")
```

```python
df["DrivingCommuterDistance"]= df["DrivingCommuterDistance"].astype("str")
df["DrivingCommuterDistance"]= df["DrivingCommuterDistance"].str.strip("-")
df
```

```python
df["DrivingCommuterDistance"]= df["DrivingCommuterDistance"].astype("int")
```

```python
df['AnnualProfessionalDevHrs'] = df['AnnualProfessionalDevHrs'].replace('?',np.nan).astype(float)
df['AnnualProfessionalDevHrs'] = df['AnnualProfessionalDevHrs'].fillna((df['AnnualProfessionalDevHrs'].median()))
mode_value = df['TextMessageOptIn'].mode()[0]
df['TextMessageOptIn'] = df['TextMessageOptIn'].fillna(mode_value)
```

```python
fig, ax = plt.subplots()
ax.boxplot(df['Age'])
q1 = np.quantile(df['Age'], 0.25)
q3 = np.quantile(df['Age'], 0.75)
iqr = q3 - q1
upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
outliers = [x for x in df['Age'] if x < lower_bound or x > upper_bound]
num_outliers = len(outliers)
print("Number of outliers:", num_outliers)
plt.show()
```
